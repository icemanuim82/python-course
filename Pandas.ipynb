```python
import pandas as pd
import numpy as np
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data , index=labels)
df
```

```python
#1-Print the three first rows using head() method
print("the three first rows")
df.head(3)
```

```python
#2-Delete rows with Nan values
df=df.dropna()
print("Delete rows with Nan values")
df
```

```python
#3-Extract the 'name' and 'score' columns from the DataFrame.
print("Select specific columns and rows:")
df[['name', 'score']]
```

```python
#4-Write a Pandas program to append a new row 'k' to data frame with these values
dict2={'name' : "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
df2 = pd.DataFrame(dict2, index=['k'])
new_df=pd.concat([df,df2])
new_df
#type(new_df)

```

```python
#5-Write a Pandas program to delete the 'attempts' column from the DataFrame.
new_df=new_df.drop("attempts",axis=1)
new_df
```
